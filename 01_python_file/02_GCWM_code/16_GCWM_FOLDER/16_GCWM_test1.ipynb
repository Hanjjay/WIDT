```python
# CSV 파일 읽어들이기

import csv  # csv 모듈 임포트
import pprint

# 파일 위치, 경로 정의
file_path = '01_South_Korea/patient.csv'

# patient.csv 파일 열기
with open(file_path) as file:
    reader = csv.reader(file)  # CSV 파일을 읽어들이는 읽기 객체
    raw = list(reader)      #  CSV 파일 내용을 리스트로 읽어들인다

pprint.pprint(raw[:10])  # 읽어들인 내용을 화면에 출력
```

```python
unknown_sex = 0
male_count = 0
female_count = 0 
for patient_count in range(len(raw[1:])):
    if raw[patient_count+1][1] == 'male':
        male_count += 1
    elif raw[patient_count+1][1] == 'female':
        female_count += 1
    else:
        unknown_sex += 1
        
print(male_count)
print(female_count)
print(unknown_sex)
print(male_count+female_count+unknown_sex)

```

```python
isolate_count = 0
released_count = 0
decased_count = 0


for patient_count in range(len(raw[1:])):
    if raw[patient_count+1][14] == 'isolated':
        isolate_count += 1
    elif raw[patient_count+1][14] == 'released':
        released_count += 1
    else:
        decased_count += 1
        
print(isolate_count)
print(released_count)
print(decased_count)
```

```python
agekid = 0
age10 = 0
age20 = 0
age30 = 0
age40 = 0
age50 = 0
age60 = 0
age70 = 0
age80 = 0
age90 = 0
ageunknown = 0



for patient_count in range(len(raw[1:])):
    if raw[patient_count+1][2] == '':
        ageunknown += 1
    else:
        check = int(raw[patient_count+1][2]) 
        if 2010 < check <= 2020:
            agekid += 1
        elif  2000 < check <= 2010:
            age10 += 1
        elif 1990 < check <= 2000:
            age20 += 1
        elif 1980 < check <= 1990:
            age30 += 1
        elif  1970 < check <= 1980:
            age40 += 1
        elif 1960 < check <= 1970:
            age50 += 1
        elif  1950 < check <= 1960:
            age60 += 1
        elif  1940 < check <= 1950:
            age70 += 1
        elif  1930 < check <= 1940:
            age80 += 1
        elif 1920 < check <= 1930:
            age90 += 1
print(agekid)
print(age10)
print(age20)
print(age30)
print(age40)
print(age50)
print(age60)
print(age70)
print(age80)
print(age90)
print(ageunknown)
```
